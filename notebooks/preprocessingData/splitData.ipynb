```python
import pandas as pd

df = pd.read_csv("featuredData.csv")

df["match_date"] = pd.to_datetime(df["match_date"], errors="coerce")

df = df.sort_values(
    ["world_cup_year", "match_date", "key_id"]
).reset_index(drop=True)

# Split por ediciones del Mundial
train_df = df[
    df["world_cup_year"] <= 2010
].copy()

validation_df = df[
    df["world_cup_year"] == 2014
].copy()

test_df = df[
    df["world_cup_year"].isin([2018, 2022])
].copy()

train_df.to_csv("train.csv", index=False, encoding="utf-8")
validation_df.to_csv("validation.csv", index=False, encoding="utf-8")
test_df.to_csv("test.csv", index=False, encoding="utf-8")

print("Train:", train_df.shape)
print("Validation:", validation_df.shape)
print("Test:", test_df.shape)

total = len(df)

print("\nPorcentajes:")
print("Train:", round(len(train_df) / total * 100, 2), "%")
print("Validation:", round(len(validation_df) / total * 100, 2), "%")
print("Test:", round(len(test_df) / total * 100, 2), "%")

print("\nAños en train:")
print(sorted(train_df["world_cup_year"].unique()))

print("\nAños en validation:")
print(sorted(validation_df["world_cup_year"].unique()))

print("\nAños en test:")
print(sorted(test_df["world_cup_year"].unique()))
```
