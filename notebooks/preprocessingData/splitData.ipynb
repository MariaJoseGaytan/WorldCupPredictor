```python
import pandas as pd

df = pd.read_csv("cleanedData.csv")
df["date"] = pd.to_datetime(df["date"])


df = df.sort_values("date").reset_index(drop=True)


# Para identificar los partidos especificos del Mundial
world_cup_name = "FifaWorldCup"

# =========================
# Creación de TRAIN
# Todos los partidos anteriores a 
# 2014 y que no sean del Mundial
# =========================

train_df = df[
    (
        (df["date"] < "2014-01-01")
    )
    |
    (
        (df["date"] >= "2014-01-01") &
        (df["date"] < "2015-01-01") &
        (df["tournament"] != world_cup_name)
    )
]

# =========================
# Creación de VALIDATION
# Mundial 2014
# =========================

validation_df = df[
    (df["date"] >= "2014-01-01") &
    (df["date"] < "2015-01-01") &
    (df["tournament"] == world_cup_name)
]

# =========================
# Creación de TEST
# Mundiales 2018 y 2022
# =========================

test_df = df[
    (
        (df["date"] >= "2018-01-01") &
        (df["date"] < "2019-01-01") &
        (df["tournament"] == world_cup_name)
    )
    |
    (
        (df["date"] >= "2022-01-01") &
        (df["date"] < "2023-01-01") &
        (df["tournament"] == world_cup_name)
    )
]


# SHUFFLE SOLO TRAIN
train_df = train_df.sample(frac=1, random_state=42).reset_index(drop=True)


# Guardar los datasets en archivos CSV
train_df.to_csv("train.csv", index=False)
validation_df.to_csv("validation.csv", index=False)
test_df.to_csv("test.csv", index=False)

print("Train:", train_df.shape)
print("Validation:", validation_df.shape)
print("Test:", test_df.shape)
```
